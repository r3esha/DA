```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
upload=files.upload()

data=pd.read_csv("Blank Quiz.csv")

participation = data['On a scale of 1 to 5, how actively do you participate in university social events and activities? (1 being not at all, 5 being very active)']
well_being = data['How would you rate your overall sense of well-being and happiness at the university on a scale of 1 to 5? (1 being very low, 5 being very high)']
social_relationships = data['To what extent do you believe your social relationships at university contribute to your overall well-being']
balance = data['How well do you think you balance your participation in university social life with academic responsibilities?']

plt.figure(figsize=(8, 6))
plt.hist(participation, bins=5, range=(1, 6), edgecolor='black')
plt.xlabel('Participation')
plt.ylabel('Count')
plt.title('Participation in University Social Events and Activities')
plt.xticks(range(1, 6))
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(well_being, bins=5, range=(1, 6), edgecolor='black')
plt.xlabel('Well-being')
plt.ylabel('Count')
plt.title('Well-being at the University')
plt.xticks(range(1, 6))
plt.show()

labels = ['Boosts Well-being', 'No Significant Impact', 'Hinders Well-being']
sizes = [
    social_relationships.str.contains('boosts', case=False).sum(),
    social_relationships.str.contains('no significant impact', case=False).sum(),
    social_relationships.str.contains('hinders', case=False).sum()
]
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Contribution of Social Relationships to Well-being')
plt.show()

labels = ['Not well at all', 'Not very well', 'Moderately well', 'Very well']
counts = balance.value_counts().reindex(labels)
plt.figure(figsize=(8, 6))
plt.bar(counts.index, counts.values)
plt.xlabel('Balance')
plt.ylabel('Count')
plt.title('Balance between Social Life and Academic Responsibilities')
plt.show()

suggestions = data['What changes or improvements, if any, would you suggest to enhance the positive impact of university social life on students\' well-being? Please select all that apply']

suggestion_counts = suggestions.str.split(';').explode().value_counts()

plt.figure(figsize=(8, 6))
plt.bar(suggestion_counts.index, suggestion_counts.values)
plt.xlabel('Suggestions')
plt.ylabel('Count')
plt.title('Suggestions to Enhance the Positive Impact of University Social Life')
plt.xticks(rotation=90)
plt.show()
```
